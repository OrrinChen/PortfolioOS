# Cost Model Demo
The MVP uses simple parametric fee and slippage formulas. This notebook shows the shape of the slippage term.

```python
import numpy as np
import pandas as pd

q = np.array([100, 500, 1000, 5000], dtype=float)
adv = 2_000_000.0
price = 60.0
k = 0.02
slippage = price * k * q * np.sqrt(q / adv)
pd.DataFrame({'shares': q, 'slippage': slippage})
```
